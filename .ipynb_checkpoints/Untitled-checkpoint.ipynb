```python
#4 counting the number of times the speed exceed speed limit
count=0
c=0
for (speed,splm) in zip( df.speed_mph, df.speedlimit_mph):
    if speed> splm:
        count=count+1
    c=c+1
print(count)
print(c)
print(count/c)
```

```python
#5 probability of exceeding speed limit when car in controlled access
c1=0
c2=0
c3=0
c4=0
c=0
for (speed,splm,controlled) in zip( df.speed_mph, df.speedlimit_mph,df.controlledAccess):
    if speed> splm:
        if controlled=='Y':
            c1=c1+1
        else:
            c2=c2+1
    else:
        if controlled=='Y':
            c3=c3+1
        else:
            c4=c4+1
    c=c+1
print(c1)
print(c2)
print(c3)
print(c4)
print(c)
print(c1/c)
print(c2/c)
print(c3/c)
print(c4/c)
#x=['overspeed_in_control', 'overspeed_not_in_control', 'speed_in_control', 'speed_not_in_control']
#y=[c1,c2,c3,c4]
```

```python
#7 histogram of speed limit
c1=0
c2=0
c3=0
c4=0
c=0
for (speed,splm,highway) in zip( df.speed_mph, df.speedlimit_mph,df.isHighway):
    if speed> splm:
        if highway==1:
            c1=c1+1
        else:
            c2=c2+1
    else:
        if highway==1:
            c3=c3+1
        else:
            c4=c4+1
    c=c+1
print(c1)
print(c2)
print(c3)
print(c4)
print(c)
print(c1/c)
print(c2/c)
print(c3/c)
print(c4/c)
print(c1/(c1+c2))
```
